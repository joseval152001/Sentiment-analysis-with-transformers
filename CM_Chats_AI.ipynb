## Confusion Matrix BARD

```python
df_bard = pd.read_excel("Test_set_BARD.xlsx")
df_bard.head()
```

```python
df_bard['Categorización'] = df_bard['Categorización'].replace({'MIXTO': 'neutral', 'BUENO': 'positive', 'MALO': 'negative'})
df_bard['Unnamed: 3'] = df_bard['Unnamed: 3'].replace({'MIXTO': 'neutral', 'POSITIVO': 'positive', 'NEGATIVO': 'negative'})
```

```python
a = df_bard["Unnamed: 3"].to_list()
b = df_bard["Categorización"].to_list()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

labels_graph = ["negative", "neutral", "positive"]

cm = confusion_matrix(b, a, labels = labels_graph)

plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=labels_graph,
                yticklabels=labels_graph)
plt.xlabel('Valores predichos', fontsize=14)
plt.ylabel('Valores verdaderos', fontsize=14)
plt.title('Matriz de Confusión', fontsize=16)
plt.show()
```
